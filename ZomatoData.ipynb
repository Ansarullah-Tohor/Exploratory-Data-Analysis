```python
df_country=pd.read_excel(r"C:\Users\mdans\Downloads\Zomatodataset\Country-Code.xlsx")
df_country.head()
```

```python
#combine df and df_country
df1=pd.merge(df,df_country,on='Country Code',how='left')
df1.head()
```

```python
# count the record from each country
country_names=df1.Country.value_counts().index #df1['Country'].value_counts()

country_values=df1.Country.value_counts().values
```

```python
ratings=df1.groupby(['Aggregate rating','Rating text','Rating color']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
# create a barplot between aggregate rating and Rating count
import matplotlib 
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
# create a barplot between aggregate rating and Rating count
import matplotlib 
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green'])
```

```python
## count plot 
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green'])
```
